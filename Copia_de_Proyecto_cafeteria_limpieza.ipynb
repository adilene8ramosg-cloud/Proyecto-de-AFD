Archivo de excel

```python
import pandas as pd

file_path = '/content/drive/MyDrive/Herramientas Datos/Coffee Shop Sales.xlsx'
df = pd.read_excel(file_path)
display(df.head())
```

Agregar columna nueva

Agrega una nueva columna que se llame state store location, en donde pongas el nombre de las sucursales, por ejemplo: las que se llaman Lower Manhattan están en "Quintana Roo", las se llaman en Astoria están en "Yucatan" y las que se llaman Hell´s Kitchen están en "Campeche"

```python
store_location_to_state = {
    'Lower Manhattan': 'Quintana Roo',
    'Astoria': 'Yucatan',
    'Hell´s Kitchen': 'Campeche'
}

df['state store location'] = df['store_location'].map(store_location_to_state)

# Display the DataFrame with the new column
display(df.head())
```

Cambia el nombre de la columna

Prompt:
Cambia el nombre de la columnas store location por store name

```python
df.rename(columns={'store_location': 'store_name'}, inplace=True)
display(df.head())
```

Producto más vendido en general

```python
most_sold_product = df.groupby('product_detail')['transaction_qty'].sum().idxmax()
most_sold_qty = df.groupby('product_detail')['transaction_qty'].sum().max()

print(f"The most sold product overall is '{most_sold_product}' with a total of {most_sold_qty} units sold.")
```

```python
output_file_path = '/content/drive/MyDrive/Herramientas Datos/Coffee Shop Sales_Modified.xlsx'
df.to_excel(output_file_path, index=False)
print(f"El archivo modificado ha sido guardado en: {output_file_path}")
```
